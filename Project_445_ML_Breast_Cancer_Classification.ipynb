Importing the Dependencies


```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column  to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and column in the dataset
data_frame.shape
```

```python
#getting some information about the data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
#statistical measure about the data
data_frame.describe()
```

```python
#checking the distribution of Target Variable
data_frame['label'].value_counts()
```

1 -- Benign
0 -- Malignant

```python
data_frame.groupby('label').mean()
```

Separating the fetures and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Decision Tree/Random Forest/Logistic Regression


```python
model = RandomForestClassifier(
    n_estimators=100,          # number of trees
    max_depth=3,            # control overfitting
    min_samples_split=2,
    random_state=42,
    n_jobs=-1                  # use all CPU cores
)

#model = DecisionTreeClassifier(
    #criterion='gini',
   # max_depth=None,         #  e.g., max_depth=5
   # random_state=42         # for reproducibility

   #model = LogisticRegression(c=100)


```

```python
# training the Decision Tree model using Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print('Accuracy on training data =', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```

```python
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {accuracy:.3f}")
```

```python
# Optional: Display a classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))


```

Building a predictive System

```python
input_data = ((20.45, 24.32, 135.10, 1300, 0.1650, 0.2450, 0.2100, 0.1100, 0.3000, 0.0950,
 0.5200, 1.2000, 4.2000, 60.50, 0.00980, 0.04000, 0.04500, 0.02000, 0.03000, 0.00550,
 24.50, 31.60, 170.20, 1850, 0.2200, 0.4300, 0.4200, 0.1900, 0.5000, 0.1500)
)

#change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The Breast cancer is Malignant')

else:
  print('The Breast Cancer is Benign')



```
